```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def quartic_function(x):
    return 2 * x**4 + 3 * x**3 - 6 * x**2 + 5 * x - 8

def plot(x ,y ,root1=None, root2=None):
    plt.plot(x, y)
    plt.title("Graph of f(x)")
    valid_roots = [r for r in [root1, root2] if r is not None]
    if valid_roots:
        plt.scatter(valid_roots, [quartic_function(r) for r in valid_roots], color='red', label='Roots', zorder=5)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True, axis='both')
    plt.ylim(-40, 50)
    plt.show()
```

```python
x = np.linspace(-4, 4)
y = quartic_function(x)
plot(x, y)
```

```python
# The Intermediate Value Theorem (IVT) guarantees that if func(a) and func(b) have opposite signs,
# there exists at least one root in (a, b) for a continuous function

def root_calculator(func, a, b, iteration = 100, tolerance=1e-6):
    if b <= a:
        a, b = b, a

    fa = func(a)
    fb = func(b)

    if fa * fb <= 0:
        print("f(a) and f(b) have different sign.\nContinue to run the program.")
        fmid = fb
    else:
        print("f(a) and f(b) cannot have the same sign.")
        return None

    for i in range(iteration):
        mid = (a + b)/2
        fmid = func(mid)

        # print(f"{i+1} loop:\na:{a} fa:{fa}\nb:{b} fb:{fb}\n")

        if abs(fmid) < tolerance:
            print(f"The function is convergence at {tolerance} using {i} loops. Returning best approximation.")
            return mid

        if fmid * fa >= 0:
            a = mid
        else:
            b = mid

    print(f"Reached max iterations ({iteration}). Returning best approximation.")
    return mid
```

```python
root1 = root_calculator(quartic_function, -3, -2)
print(root1)
```

```python
root2 = root_calculator(quartic_function, 1, 2)
print(root2)
```

```python
plot(x, y, root1, root2)
```
